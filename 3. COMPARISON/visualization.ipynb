```
import pandas as pd
%pylab inline
plt.rcParams.update({'font.size': 16})
```

```
base_map = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/MAP.csv')
#kernel_plot = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data_spacy/MAP_kernel_plot.csv')
```

```
fig, ax = plt.subplots(1)
#fig.set_size_inches(10, 4)
base_map[0::8].plot.barh(x='embedding', y=['MAP', 'recall'], ax=ax)
ax.set_ylabel('Embedding')
ax.set_title('MAP and Recall on dataset 100 with different models')
```

```
base_mean = base_map.groupby('train size').mean().reset_index()
base_mean.plot(x='train size', y=['recall', 'MAP'])
plt.title('Mean recall and precision on datasets with different size')
```
